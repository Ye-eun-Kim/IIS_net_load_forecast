```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./processed_data/netload/RISE_all_interpolated.csv',index_col=0, header=0)
display(df)

```

```python
# plot graph every column of dataframe
# check 210111-210124


plt.xticks = plt.xticks


df_1 = df.iloc[10:24, :]
date_list = np.array(df_1.index).astype(str)


for i in range(0, 24):
    print('This is the ', i, 'th graph')
    fig = plt.figure(figsize=(10, 2.5))
    ax = plt.axes()
    col = df_1.iloc[:,i]
    ax.plot(date_list, col)
    ax.set_xticks(range(len(date_list)))
    ax.set_xticklabels(date_list, rotation = 45)
    plt.show()
```
